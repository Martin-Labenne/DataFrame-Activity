```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Ouverture de 60s, 70s et 80s

```python
sixties = pd.read_csv('60s.csv', index_col = 0)
seventies = pd.read_csv('70s.csv', index_col = 0)
heighties = pd.read_csv('80s.csv', index_col = 0)
```

# Graph

```python
plt.scatter(sixties.index, sixties['births'])
plt.scatter(seventies.index, seventies['births'])
plt.scatter(heighties.index, heighties['births'])
plt.show()
```
